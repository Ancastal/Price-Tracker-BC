```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine("sqlite:///transactions.db", echo=False)
df = pd.read_sql("SELECT * FROM transactions", con=engine)
```

```python
import matplotlib.pyplot as plt

# Convert the 'timestamp' column to datetime type
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Group the data by 'timestamp' and calculate the mean price for each date
mean_prices = df.groupby('timestamp')['price'].mean()

# Plot the mean prices over time
plt.plot(mean_prices.index, mean_prices.values)

# Add title and labels
plt.title('Redstone Price Change Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')

# Show the chart
plt.show()
```

```python
# get the most common price for redstone with 10 quantity
quantity = 16
try:
    most_common_price = df[(df["item_name"] == "Redstone") & (df["quantity"] == quantity)]["price"].mode()[0]
except:
    most_common_price = df[(df["item_name"] == "Redstone")]["price"].div(df["quantity"]).median()
    most_common_price *= quantity    
print(most_common_price)
```

```python
import matplotlib.pyplot as plt

redstone_df = df[df['item_name'] == 'Redstone']

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(redstone_df['timestamp'], redstone_df['price'], color='red')
ax.set_xlabel('Date')
ax.set_ylabel('Price (coins)')
ax.set_title('Redstone Price over Time')
plt.show()
```
